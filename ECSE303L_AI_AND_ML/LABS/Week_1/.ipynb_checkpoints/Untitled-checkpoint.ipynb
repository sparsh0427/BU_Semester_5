```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
data = pd.read_csv('wine.data', names=['class', 'Alcohol','Malic acid','Ash','Alcalinity of ash', 
 	 'Magnesium',
	 'Total phenols',
 	 'Flavanoids',
 	 'Nonflavanoid phenols',
 	 'Proanthocyanins',
	'Color intensity',
 	'Hue',
 	'OD280/OD315 of diluted wines',
   'Proline'])
print("FOR .DATA EXTENSION --> ")
print()
print("FILE DATA = ")
print(data)
print()
print("SIZE OF MATRIX = ",data.size)
print("SHAPE OF MATRIX = ", data.shape)
print("NUMBER OF ROWS IN THE MATRIX = ", len(data))
print("NUMBER OF COLUMNS IN THE MATRIX = ", len(data.columns))
print("TRANSPOSING MATRIX = ",data.transpose())
print("INVERSION = ",np.linalg.pinv(data))
print("DOT PRODUCT = ",data.dot(data.transpose()))

# LINE GRAPH 
# data.drop(['class'],axis=1).plot.line(title="Wine dataset")
# data.plot(x="Alcohol", y="Ash")


# Bar Graph
# plt.bar("Alcohol", "Malic acid", data = data, color = "red")
# plt.xlabel("Alcohol")
# plt.ylabel("Ash")
# plt.show()

```
