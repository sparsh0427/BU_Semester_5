#### Extension --> .data

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
data = pd.read_csv('wine.data', names=['class', 'Alcohol','Malic acid','Ash','Alcalinity of ash', 
 	 'Magnesium',
	 'Total phenols',
 	 'Flavanoids',
 	 'Nonflavanoid phenols',
 	 'Proanthocyanins',
	'Color intensity',
 	'Hue',
 	'OD280/OD315 of diluted wines',
   'Proline'])
print("FOR .DATA EXTENSION --> ")
print()
print("FILE DATA = ")
print(data)
print()
print("SIZE OF MATRIX = ",data.size)
print()
print("SHAPE OF MATRIX = ", data.shape)
print()
print("NUMBER OF ROWS IN THE MATRIX = ", len(data))
print()
print("NUMBER OF COLUMNS IN THE MATRIX = ", len(data.columns))
print()
print("TRANSPOSING MATRIX = ")
print()
print(data.transpose())
print()
print("INVERSION = ",np.linalg.pinv(data))
print()
print("DOT PRODUCT = ",data.dot(data.transpose()))
print()

```

```python
import cv2
import numpy as np
import sys
import webcolors
import time

cam=cv2.VideoCapture('video.avi')
_, fo = cam.read()
framei = cv2.cvtColor(fo, cv2.COLOR_BGR2GRAY)
bg_avg = np.float32(framei)
video_width = int(cam.get(3))
video_height = int(cam.get(4))
fr = int(cam.get(5))
print("frame rate of stored video:::",fr)

while(cam.isOpened): 
    
    if cv2.waitKey(25) & 0xFF == ord('q'): 
        break

    f,img=cam.read()
    start_fps=time.time()
    k = cv2.waitKey(20)
    if(k == 27):
        break
    endtime_fps=time.time()
    diff_fps=endtime_fps-start_fps
    print("Frame rate::",1/diff_fps)
```

##### MFCC ( Mel-Frequency Cepstral Coefficients )

```python
x, sr = librosa.load('audio.wav')
mfccs = librosa.feature.mfcc(x, sr)
print(mfccs.shape)
#Displaying  the MFCCs:
librosa.display.specshow(mfccs, sr=sr, x_axis='time')
```

##### STFT ( Short term fourier transform )

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz') 
#If to pring log of frequencies  
#librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()
```

#### Wordcloud

```python
# Python program to generate WordCloud 

# importing all necessery modules 
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd 

# Reads 'Youtube04-Eminem.csv' file 
df = pd.read_csv(r"wordcloud.csv", encoding ="latin-1") 

comment_words = '' 
stopwords = set(STOPWORDS) 

# iterate through the csv file 
for val in df.CONTENT: 
	
	# typecaste each val to string 
	val = str(val) 

	# split the value 
	tokens = val.split() 
	
	# Converts each token into lowercase 
	for i in range(len(tokens)): 
		tokens[i] = tokens[i].lower() 
	
	comment_words += " ".join(tokens)+" "

wordcloud = WordCloud(width = 800, height = 800, 
				background_color ='white', 
				stopwords = stopwords, 
				min_font_size = 10).generate(comment_words) 

# plot the WordCloud image					 
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 

plt.show() 
```
