#### Matrix related operations such as transpose, inversion, dot product (with transpose of same matrix)

#### Extension --> .data

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
data = pd.read_csv('wine.data', names=['class', 'Alcohol','Malic acid','Ash','Alcalinity of ash', 
 	 'Magnesium',
	 'Total phenols',
 	 'Flavanoids',
 	 'Nonflavanoid phenols',
 	 'Proanthocyanins',
	'Color intensity',
 	'Hue',
 	'OD280/OD315 of diluted wines',
   'Proline'])
print("FOR .DATA EXTENSION --> ")
print()
print("FILE DATA = ")
print(data)
print()
print("SIZE OF MATRIX = ",data.size)
print()
print("SHAPE OF MATRIX = ", data.shape)
print()
print("NUMBER OF ROWS IN THE MATRIX = ", len(data))
print()
print("NUMBER OF COLUMNS IN THE MATRIX = ", len(data.columns))
print()
print("TRANSPOSING MATRIX = ")
print()
print(data.transpose())
print()
print("INVERSION = ",np.linalg.pinv(data))
print()
print("DOT PRODUCT = ",data.dot(data.transpose()))
print()

```

#### LINE GRAPH

```python
data[['Ash','Malic acid']].plot.line(title="Wine dataset", figsize=(7,7))
```

#### BAR GRAPH

```python
data_sliced = data.iloc[2:6]
data_sliced[['Ash','Malic acid']].plot.bar()
```

#### HISTOGRAM

```python

data[['Ash']].plot.hist()
```

#### PIE CHART

```python
data_trans = data.transpose()
data_trans.drop(['class','Proline'],axis=0).plot.pie(y=0,figsize=(10,10))
```

#### SCATTER PLOT

```python
data.plot.scatter(x='Alcohol', y='Ash')
```
