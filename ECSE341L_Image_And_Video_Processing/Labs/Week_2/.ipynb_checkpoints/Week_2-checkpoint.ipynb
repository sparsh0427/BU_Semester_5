## 1. RGB to Gray-scale conversion of images


### a) Read any RGB image as input and convert it into grayscale image using direct function/library in Python. Show input and output together using subplot function

```python
from PIL import Image
img = Image.open("img.jpg")
img_grey = img.convert('LA')
```

```python
import matplotlib.pyplot as plt
plt.subplot(211)
plt.imshow(img)
plt.subplot(212)
plt.imshow(img_grey)
```

### b) RGB to grayscale Image conversion using numpy and matplotlib only. No other special Python library can be used. Explain the steps in your own words

```python
import numpy as np
import matplotlib as pyplot
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

an_image = mpimg.imread(r"img.jpg")

rgb_weights = [0.2989, 0.5870, 0.1140]
grayscale_image = np.dot(an_image[...,:3], rgb_weights)


plt.subplot(211)
plt.imshow(an_image)
plt.subplot(212)
plt.imshow(grayscale_image,cmap="gray")
```

## 2. Take two images and perform Image subtraction, addition, and multiplication of those images using arrays. 

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
img = mpimg.imread('img.jpg') #This will give array
img1 = mpimg.imread('img2.jpg') #This will give array

```

#### Image Addition

```python
img_add = img + img1
plt.imshow(img_add)
```

#### Image Subtraction

```python
img_sub = img - img1
plt.imshow(img_sub)
```

#### Image Multiplication

```python
img_mul = img * img1
plt.imshow(img_mul)
```

## 3. Converting from one color space into another (RGB to HSV conversion)


```python
import cv2 
HSV_img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cv2.namedWindow("Display 1")
cv2.resizeWindow("Display 1", 300, 300)
cv2.imshow("Display 1", HSV_img)
cv2.imwrite('hsv.jpg',HSV_img) 
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 4. Image resizing using bilinear, bicubic and nearest neighbour interpolation methods. What is your observation about the outputs of these methods; can you see any difference in outputs? Write it in your own words.

```python
import cv2
 
img = cv2.imread(r"c1.jpg", cv2.IMREAD_UNCHANGED)
 
print('Original Dimensions : ',img.shape)
 
scale_percent = 80 # percent of original size
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)

# resize image
#bilinear
resized_bi = cv2.resize(img, (500,500), interpolation = cv2.INTER_LINEAR)
#nearest neighbour
resized_nn = cv2.resize(img, (500,500), interpolation = cv2.INTER_NEAREST)
#bicubic
resized_bc = cv2.resize(img, (500,500), interpolation = cv2.INTER_CUBIC)

cv2.imshow("Original Image", img)
cv2.waitKey(0)
cv2.imshow("Bilinear Resize", resized_bi)
cv2.imwrite("Bilinear Resize.jpg", resized_bi)
cv2.waitKey(0)
cv2.imshow("Nearest Neighbour Resize", resized_nn)
cv2.imwrite("Nearest Neighbour Resize.jpg", resized_nn)
cv2.waitKey(0)
cv2.imshow("Bicubic Resize", resized_bc)
cv2.imwrite("Bicubic Resize.jpg", resized_bc)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
