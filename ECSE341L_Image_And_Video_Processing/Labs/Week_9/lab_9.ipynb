# Q1

```python
import matplotlib.pyplot as plt 

import numpy as np

from skimage.io import imread, imsave 

from skimage.morphology import erosion, dilation, opening, closing, white_tophat 

from skimage.morphology import black_tophat, skeletonize, convex_hull_image 

from skimage.morphology import disk, square, diamond 

from skimage.color import rgb2gray 

from skimage.util import img_as_ubyte 

im = imread('j.png', as_gray=True) 
im = rgb2gray(im) 

selem = square(5)

eroded = erosion(im, selem)

req = np.subtract(im, eroded)



plt.figure(figsize=(6,6))
plt.subplot(121), plt.imshow(im, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(req, cmap='gray'),plt.title('Boundary')
plt.xticks([]), plt.yticks([])
plt.show()
```

# Q3

```python
from skimage.morphology import binary_erosion, rectangle
im = rgb2gray(imread('j.png'))
print(np.max(im))
im[im <= 0.5] = 0
im[im > 0.5] = 1

se = rectangle(1,5)
im1 = binary_erosion(im, se)


se = rectangle(1,15)
im2 = binary_erosion(im, se)



plt.figure(figsize=(10,10))
plt.subplot(131), plt.imshow(im, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(im1, cmap='gray'),plt.title('erosion with rectangle size (1,5) ')
plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(im2, cmap='gray'),plt.title(' erosion with rectangle size (1,15)')
plt.xticks([]), plt.yticks([])
plt.show()
```
