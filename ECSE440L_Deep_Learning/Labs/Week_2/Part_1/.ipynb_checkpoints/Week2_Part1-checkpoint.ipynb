#### b. Create an empty DataFrame with columns name only then appending rows one by one to it using append() method. Empty data frame should be having 3 columns “student name”,”date of birth”, “marks in english”. Append one by one 5 random rows to this dataframe.


```python
data = pd.DataFrame(columns=['Student Name','Date Of Birth','Marks in Maths'])
listOfSeries = [pd.Series(['Sparsh', '27/04/2000', 95], index=data.columns ) ,
                pd.Series(['Anurup', '01/05/2001', 94], index=data.columns ) ,
                pd.Series(['Amit', '01/01/1999',94], index=data.columns ),
               pd.Series(['Shubh', '12/05/2000',85], index=data.columns ),
               pd.Series(['Tejas', '11/11/1989',82], index=data.columns )]
data_mod = data.append(listOfSeries , ignore_index=True)
data_mod
```

### 3. Working with series:

#### a. Create two panda series of 100 random numbers


```python
data = np.random.randint(1, 100, size=100)
series_1 = pd.Series(data)
data = np.random.randint(1, 100, size=100)
series_2 = pd.Series(data)
print("Series 1 = ")
print(series_1)
print()
print("Series 2 = ")
print(series_2)
```

#### b. Write a Pandas program to add, subtract, multiply and divide two Pandas Series

```python
series_add = series_1.add(series_2)
print("Addition = ",series_add)
series_sub = series_1.sub(series_2)
print("Subtraction = ",series_sub)
series_mul = series_1.mul(series_2)
print("Multiplication = ",series_mul)
series_div = series_1.div(series_2)
print("Division = ",series_div)
```

####  c. Store all the 6 series to a dataframe with column names as “number 1” “number 2”, “add”, “sub”, “mult” and “div”


```python
data ={'Number 1':series_1,'Number 2':series_2,'Add':series_add,'Sub':series_sub,'Mult':series_mul,'Div':series_div} 
data_series = pd.DataFrame(data)
data_series
```
