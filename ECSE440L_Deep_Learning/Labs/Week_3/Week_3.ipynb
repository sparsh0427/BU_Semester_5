```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import time
```

```python
exec_time = []
train_acc = []
test_acc = []
```

```python
def modeling(size, no_of_images):
    start_time = time.time()
    x = []
    y = []
    folder = 'dogs/'
    for i in range(no_of_images):
        path = folder + 'dog.' + str(i+1) + '.jpg'
        img=plt.imread(path)
        img=cv2.resize(img,(size,size))
        img=img.flatten()
        x.append(img)
        y.append(0)  # Dog Class = 0
        
    folder = 'cats/'
    for i in range(no_of_images):
        path = folder + 'cat.' + str(i+1) + '.jpg'
        img=plt.imread(path)
        img=cv2.resize(img,(size,size))
        img=img.flatten()
        x.append(img)
        y.append(1) # Dog Class = 1

    #Converting to numy arrays - 
    x=np.array(x)
    y=np.array(y)
    
    x_train, x_test, y_train,y_test = train_test_split(x,y,train_size=0.75)
    model=LogisticRegression()
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    sklearn_test_accuracy = accuracy_score(y_test, y_pred)
    print(f"\nAccuracy in testing set by sklearn model: {sklearn_test_accuracy}")
    test_acc.append(sklearn_test_accuracy)
    y_pred = model.predict(x_train)
    sklearn_train_accuracy = accuracy_score(y_train, y_pred)
    print(f"\nAccuracy in training set by sklearn model: {sklearn_train_accuracy}")
    train_acc.append(sklearn_train_accuracy)
    print()
    end_time = time.time()
    print("Time For Execution : ",end_time-start_time)
    exec_time.append(end_time-start_time)
```

### MODEL 1

```python
modeling(500,80)
```

### MODEL 2

```python
modeling(100,100)
```

### MODEL 3 

```python
modeling(50,10)
```

### MODEL 4

```python
modeling(25,30)
```

### MODEL 5

```python
modeling(10,15)
```

## Plotting and Visualizing Data

```python
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8)) 
   
br1 = np.arange(5) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 
   
# Make the plot 
a = plt.bar(br1, exec_time, color ='r', width = barWidth,edgecolor ='grey',label='execution_time') 
b = plt.bar(br2, test_acc, color ='g', width = barWidth, 
        edgecolor ='grey', label ='test_accuracy') 
c = plt.bar(br3, train_acc, color ='b', width = barWidth, 
        edgecolor ='grey', label ='train_accuracy') 

plt.xticks([r + barWidth for r in range(len(exec_time))], 
           ['Model 1','Model 2','Model 4','Model 4','Model 5']) 
   
plt.legend(loc='best')
plt.show()
```
