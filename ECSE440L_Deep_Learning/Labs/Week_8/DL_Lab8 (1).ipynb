Eg of using vgg16 model as feature extraction model

```python
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.applications.vgg16 import preprocess_input
from keras.applications.vgg16 import decode_predictions
from keras.applications.vgg16 import VGG16
from keras.models import Model
```

```python
import numpy as np
image=cv2.imread("/content/drive/My Drive/dl assnmt 3/dogs/dog.1.jpg")
image=cv2.resize(image,(224,224))
image=np.array(image)
```

```python
image=image.reshape((1,image.shape[0],image.shape[1],image.shape[2]))
```

Prepare image for VGG model

```python
image=preprocess_input(image)
```

Predict probability across all output classes

```python
model=VGG16()
yhat=model.predict(image)
label=decode_predictions(yhat)
#highest prob
label=label[0][0]
#print classification
print('%s(%.2f%%' % (label[1],label[2]*100))
```

```python
vgg_conv = VGG16(weights='imagenet' , include_top=False, input_shape=(224,224,3))
```

Freezing

```python
#freeze all layer
for layer in vgg_conv.layers[:]:
  layer.trainable=False
#individual layers
for layer in vgg_conv.layers:
  print(layer,layer.trainable)
```

```python
from tensorflow.keras.layers import Dense , Dropout , Flatten
from keras.models import Sequential
#creating model
model=Sequential()

#adding vgg conv base model

model.add(vgg_conv)
model.add(Flatten())
model.add(Dense(1024,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(2,activation='softmax'))
model.summary()
```
